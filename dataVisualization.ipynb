```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
example = cv2.imread("data_example/adwp/B02.tif", cv2.IMREAD_UNCHANGED)
plt.imshow(example)
```

```python
features = "data_example/adwp/"
for feat in os.listdir(features):
    path = os.path.join(features, feat)
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    plt.figure()
    plt.imshow(img, cmap='gray')
```

```python
img.shape, np.max(img), np.min(img)
```

```python
# Label visualization
label = cv2.imread("data_example/adwp.tif", cv2.IMREAD_UNCHANGED)
print(np.unique(label), label.shape)
plt.imshow(label * 255)
```
